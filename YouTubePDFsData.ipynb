## Youtube

```python
import os
from langchain_groq import ChatGroq
from langchain.prompts import ChatPromptTemplate
from langchain_community.document_loaders import YoutubeLoader
```

```python
# API key for Groq
api_key = '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'
```

```python
# Define the key as an enviroment variable
os.environ['GROQ_API_KEY'] = api_key # Use your Groq API key here

# Create a chat model instance
chat = ChatGroq(model='meta-llama/llama-4-scout-17b-16e-instruct')
```

```python
url = 'https://youtu.be/HuzdCvxkToc?si=brEbN7BYBRDBfep9'

loader = YoutubeLoader.from_youtube_url(url, language='pt')
docs = loader.load()
```

```python
# Puting the content in a string
docs_list = str([doc.page_content for doc in docs][0])[:-4000] # Trimming the strin becaue the model has a limit of tokens
docs_list
```

```python
# Creating the template
template = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful assistant called Jarvis and answer questions about a Youtube video where the subtitle is: {subtitle}'),
    ('user', '{input}')
])

chain = template | chat
```

```python
# invoking the model with our subtitle and the prompt
answer = chain.invoke({
    'subtitle': docs_list,
    'input': 'Faca um resumo do video'
})
print(answer.content)
```

## PDFs

```python
from langchain_community.document_loaders import PyPDFLoader
```

```python
# Load file
loader = PyPDFLoader('./De Alunos Para Alunos - Lógica Aplicada a Linguagem Programação.pdf')
list_docs = loader.load()

#
docs_list = [doc.page_content for doc in list_docs]
```

```python
print(docs_list)
```

```python
# Creating the template
template = ChatPromptTemplate.from_messages([
    ('system', 'You are a helpful assistant called Jarvis and answer questions about a Pdf document where the content is: {content}'),
    ('user', '{input}')
])

chain = template | chat

# invoking the model with our subtitle and the prompt
answer = chain.invoke({
    'content': docs_list,
    'input': 'Faca um resumo do PDF'
})
print(answer.content)
```
